```python
from sklearn.preprocessing import OneHotEncoder
import xgboost as xgb
import numpy as np
import pandas as pd
from sklearn.preprocessing import Imputer
d=pd.read_csv('./chronic_kidney_with_class.csv')
d.head()
```

```python
d.replace('?',np.nan,inplace=True)
d.head()
```

```python
d.columns
```

```python
d[['age','bp','sg','al','su','pot','hemo','pcv','wc','rc','rbc','pc','pcc','ba']]=d[['age','bp','sg','al','su','pot','hemo','pcv','wc','rc','rbc','pc','pcc','ba']].astype(float)
```

```python
d.head()
```

```python
d.shape
```

```python
d=pd.get_dummies(d)
```

```python
d.head()
```

```python
d.shape
```

```python
d.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
x=d.iloc[:,:-1]
y=d.iloc[:,-1]
imp=Imputer(missing_values='NaN',strategy='median',axis=0)
xgc=xgb.XGBClassifier(objective='binary:logistic',n_estimators=10,seed=123)
s=[('imp',imp),('xgc',xgc)]
x_tr,x_ts,y_tr,y_ts=train_test_split(x,y,test_size=0.3,random_state=2)
p=Pipeline(s)
p.fit(x_tr,y_tr)
p.predict(x_ts)
p.score(x_ts,y_ts)
```
