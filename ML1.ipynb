```python
import pandas as pd
import numpy as np
points=np.genfromtxt('boston_housing.csv',delimiter=',',skip_header=1,usecols=[0,3])
print(points)
```

```python
N=len(points)
def grad(b_cur,m_cur,learnrate,N,points):
    b_grad=0
    m_grad=0
    for i in range(0,N):
        b_grad -=(2/N)*(points[i,1]-m_cur*points[i,0]-b_cur)
        m_grad -=(2/N)*(points[i,1]-m_cur*points[i,0]-b_cur)*points[i,0]
    b_new=b_cur-(learnrate*b_grad/i)
    m_new=m_cur-(learnrate*b_grad/i)
    return b_new,m_new
```

```python
def err(b_cur,m_cur,points,error):
    e=0
    for i in range(0,N):
        e +=(points[i,1]-m_cur*points[i,0]-b_cur)**2/N
    error.append(e)
```

```python
learnrate=.1
b=4
m=1
error=[]
for i in range(0,100):
    b,m=grad(b,m,learnrate,N,points)
    err(b,m,points,error)
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(points[:,0],points[:,1],c='green')
y=[]
for i in range(0,N):
    y.append(m*points[i,0]+b)
plt.plot(points[:,0],y)
```
