```python
from sklearn.preprocessing import Imputer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
import xgboost as xgb
from sklearn.pipeline import Pipeline
```

```python
imp1=Imputer(missing_values='NaN',strategy='mean',axis=0)
imp2=Imputer(missing_values='NaN',strategy='median',axis=0)
imp3=Imputer(missing_values='NaN',strategy='most_frequent',axis=0)
dt=DecisionTreeClassifier()
xgc=xgb.sklearn.XGBClassifier(objective='binary:logistic',n_estimators=10,seed=123)
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
s1=[('imp1',imp1),('knn',knn)]
s2=[('imp2',imp2),('knn',knn)]
s3=[('imp3',imp3),('knn',knn)]
s4=[('imp1',imp1),('dt',dt)]
s5=[('imp2',imp2),('dt',dt)]
s6=[('imp3',imp3),('dt',dt)]
s7=[('imp1',imp1),('xgc',xgc)]
s8=[('imp2',imp2),('xgc',xgc)]
s9=[('imp3',imp3),('xgc',xgc)]
p1=Pipeline(s1)
p2=Pipeline(s2)
p3=Pipeline(s3)
p4=Pipeline(s4)
p5=Pipeline(s5)
p6=Pipeline(s6)
p7=Pipeline(s7)
p8=Pipeline(s8)
p9=Pipeline(s9)
```

```python
scores=[]
from sklearn.model_selection import train_test_split
x_tr,d_ts,y_tr,d_sc=train_test_split(x,y,test_size=0.3,random_state=42)
imptr=Imputer(missing_values='NaN',strategy='most_frequent',axis=0)
imptr.fit_transform(d_ts)
p1.fit(x_tr,y_tr)
p1.predict(d_ts)
scores.append(p1.score(d_ts,d_sc))

p2.fit(x_tr,y_tr)
p2.predict(d_ts)
scores.append(p2.score(d_ts,d_sc))

p3.fit(x_tr,y_tr)
p3.predict(d_ts)
scores.append(p3.score(d_ts,d_sc))

p4.fit(x_tr,y_tr)
p4.predict(d_ts)
scores.append(p4.score(d_ts,d_sc))

p5.fit(x_tr,y_tr)
p5.predict(d_ts)
scores.append(p5.score(d_ts,d_sc))

p6.fit(x_tr,y_tr)
p6.predict(d_ts)
scores.append(p6.score(d_ts,d_sc))

p7.fit(x_tr,y_tr)
p7.predict(d_ts)
scores.append(p7.score(d_ts,d_sc))

p8.fit(x_tr,y_tr)
p8.predict(d_ts)
scores.append(p8.score(d_ts,d_sc))

p9.fit(x_tr,y_tr)
p9.predict(d_ts)
scores.append(p9.score(d_ts,d_sc))
scores
```

```python
strat=['mean','most_frequent','median']
imp=[]
for i in strat:
    imp.append(('imp',Imputer(missing_values='NaN',strategy=i,axis=0)))
print(imp)
clf=[]
clf.append(('dt',DecisionTreeClassifier()))
clf.append(('knn',KNeighborsClassifier(n_neighbors=5)))
clf.append(('xgc',xgb.sklearn.XGBClassifier(objective='binary:logistics',n_estimators=10,seed=123)))
clf
```
